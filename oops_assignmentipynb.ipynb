6. Explain the concept of inheritence in oop.

 - Inheritance in OOP means that one class (child class) can inherit properties and behaviors (data and methods) from another class (parent class).
 - It allows code reusability and helps create a hierarchical relationship between classes.

Example:
 - Think of a Parent and Child:
 - The child inherits traits like eyes, hair color, etc., from the parent.

7. What is polymorphism in OOP?

 10. What are class and static method in python?

  - In Python, both class methods and static methods are special kinds of methods that belong to a class, not just its objects — but they behave differently.

   1. Class Method

  - A class method is a method that works with the class itself rather than   individual objects.

 ➤ Defined using: @classmethod decorator.
 ➤ The first parameter is always cls, which refers to the class (not the object).
   
   2. Static Method

A static method is a method that does not depend on either the class or instance.
It behaves like a regular function, but it belongs to the class’s namespace.

➤ Defined using: @staticmethod decorator
➤ No self or cls parameter is used.


```python
class Student:
    school_name = "ABC Public School"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name  # modifies class variable

# Using class method
Student.change_school("XYZ International School")

# Accessing updated class variable
print(Student.school_name)   # Output: XYZ International School
```

```python
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

# Using static method
print(MathUtils.add(5, 10))   # Output: 15
```

 12. What is method overriding in oop?

  - Method Overriding in OOP

  - Method overriding happens when a child class (subclass) defines a method  with the same name and parameters as a method in its parent class (superclass)
  - The child class’s method replaces (overrides) the parent’s version when called from an object of the child class.

 Purpose:
  To allow a subclass to modify or extend the behavior of a method defined in the parent class.

  Example:

```python
class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def sound(self):     # overriding parent method
        print("Bark Bark")

# Creating objects
a = Animal()
d = Dog()

a.sound()   # Output: Some generic animal sound
d.sound()   # Output: Bark Bark
```

13. What is property decorator in python?
  
  - Property Decorator in Python
    The property decorator in Python is used to turn a method into an attribute, allowing controlled access (get, set, delete) to class variables without changing how you access them.

  - Why Use It?
  - It’s used when you want to:
  - Hide private variables (data encapsulation)
  - Add validation or logic when getting or setting a value.
  - Access values like attributes, but with the safety of methods.

```python
Example Without @property

class Student:
    def __init__(self, name):
        self._name = name  # underscore means "private" by convention

    def get_name(self):
        return self._name

    def set_name(self, value):
        if len(value) > 0:
            self._name = value
        else:
            print("Name cannot be empty")

# Using methods
s = Student("Richa")
print(s.get_name())   # Output: Richa
s.set_name("Anita")
```

```python
Same Example Using @property

class Student:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):          # getter
        return self._name

    @name.setter
    def name(self, value):   # setter
        if len(value) > 0:
            self._name = value
        else:
            print("Name cannot be empty")

# Using property like an attribute
s = Student("Richa")
print(s.name)      # calls getter → Output: Richa

s.name = "Anita"   # calls setter
print(s.name)      # Output: Anita
```

18. What is multiple inheritence in python?

  - Multiple inheritance is a feature in object-oriented programming (OOP) where a child class inherits from more than one parent class.
  - It means a class can use properties and methods of multiple base classes.

  Example:

```python
class Father:
    def skills(self):
        print("Father: Cooking")

class Mother:
    def skills(self):
        print("Mother: Painting")

class Child(Father, Mother):
    def skills(self):
        print("Child: Dancing")
        Father.skills(self)   # explicitly calling parent method
        Mother.skills(self)

c = Child()
c.skills()


Output:
Child: Dancing
Father: Cooking
Mother: Painting
```

20. What is the significance of the super()' function in python?

  - Significance of super() Function in Python
  
  - The super() function is used to call methods of a parent (super) class from within a child (subclass).
  - It allows a subclass to inherit and extend the behavior of its parent class without rewriting code.
  
  - Purpose of super()

     super() helps in:
  - Accessing parent class methods or constructors (__init__())
  - Avoiding code duplication
  - Supporting multiple inheritance through Method Resolution Order (MRO)
  - Making code more maintainable and easier to update


```python
Basic Syntax
super().method_name()


or inside a constructor:

super().__init__()

Example 1: Calling Parent Constructor

class Parent:
    def __init__(self, name):
        self.name = name
        print("Parent constructor called")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)   # calling Parent's __init__()
        self.age = age
        print("Child constructor called")

c = Child("Richa", 21)


Output:

Parent constructor called
Child constructor called
```

22. What is the difference between @staticmethod and @classmethod in python?

  - 1. @staticmethod
  - Does not take self or cls as the first argument.
  - Cannot access instance variables or class variables directly.
  - Acts like a regular function that happens to reside in the class.
  - Can be called using class name or instance.

  Example


```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(5, 10))   # Output: 15
obj = Math()
print(obj.add(2, 3))     # Output: 5
```

2. @classmethod

   - Takes cls (class itself) as the first argument.
   - Can access or modify class variables, but cannot access instance variables.
   - Usually used for factory methods or methods that affect the class.
   - Can be called using class name or instance.

   Example:

```python
class Employee:
    company = "ABC Corp"

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

print(Employee.company)  # Output: ABC Corp
Employee.change_company("XYZ Inc")
print(Employee.company)  # Output: XYZ Inc
```

23. How does polymorphism work in python with inheritance?

  - Polymorphism with Inheritance in Python

  - Polymorphism means “many forms” — the same method or operation behaves   differently depending on the object calling it.
  - When combined with inheritance, it allows child classes to modify or extend methods from parent classes while keeping the same interface.
  

```python
Example: Animal Sounds
class Animal:
    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Bark Bark")

class Cat(Animal):
    def make_sound(self):
        print("Meow Meow")

# Using polymorphism
animals = [Dog(), Cat(), Animal()]

for animal in animals:
    animal.make_sound()


Output:

Bark Bark
Meow Meow
Some generic animal sound


Explanation:

All objects respond to the same method name (make_sound) differently.

Python determines at runtime which method to call based on the object type.

 Example with Function Using Parent Type
def animal_sound(animal: Animal):
    animal.make_sound()  # Calls appropriate method based on object

animal_sound(Dog())  # Output: Bark Bark
animal_sound(Cat())  # Output: Meow Meow


Here, one function works with any subclass of Animal — that’s polymorphism in action.
```

24. What is method chaining in python OOP?

   - Method Chaining in Python OOP
   - Method chaining is a technique where multiple methods are called on the    same object in a single statement, one after another.

   - In Python OOP, this is typically done by having methods return self (the   object itself), so that subsequent methods can be called on the same instance.

   Purpose
   - Improves code readability by reducing repeated object references.
   - Allows succinct, fluent syntax.
   - Useful for configuring objects or performing sequential operations.

   *How It Works
   
   - Each method performs an operation on the object.
   - Each method returns the object itself (self).
   - The next method can then be called directly on the returned object.

```python
Example
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)
        return self       # return self for chaining

    def display_scores(self):
        print(f"{self.name}'s scores: {self.scores}")
        return self       # return self for chaining

# Method chaining
s = Student("Richa")
s.add_score(95).add_score(88).display_scores().add_score(76).display_scores()


Output:

Richa's scores: [95, 88]
Richa's scores: [95, 88, 76]
```

25. What is the purpose of the __call__ method in python?

   - Purpose of the __call__() Method in Python
   - The __call__() method is a special (dunder) method in Python that allows an object to be called like a function.
   - If a class defines __call__(), you can use its instances as callable objects, just like functions.

   - Purpose

   - Make objects callable.
   - Enable function-like behavior while keeping object state.
   - Useful in design patterns like decorators, callbacks, or function wrappers.
   

```python
Syntax
class ClassName:
    def __call__(self, *args, **kwargs):
        # define behavior when object is called
        pass

 Example
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return self.factor * value

# Create object
double = Multiplier(2)

# Call object like a function
print(double(5))  # Output: 10
print(double(10)) # Output: 20


Explanation:

double is an object of Multiplier.

Calling double(5) internally calls double.__call__(5).
```
