  - Method overriding happens when a child class (subclass) defines a method  with the same name and parameters as a method in its parent class (superclass)

 Purpose:

13. What is property decorator in python?
  
  - Property Decorator in Python
    The property decorator in Python is used to turn a method into an attribute, allowing controlled access (get, set, delete) to class variables without changing how you access them.

  - Why Use It?
  - It’s used when you want to:
  - Hide private variables (data encapsulation)
  - Add validation or logic when getting or setting a value.
  - Access values like attributes, but with the safety of methods.

```python
Example Without @property

class Student:
    def __init__(self, name):
        self._name = name  # underscore means "private" by convention

    def get_name(self):
        return self._name

    def set_name(self, value):
        if len(value) > 0:
            self._name = value
        else:
            print("Name cannot be empty")

# Using methods
s = Student("Richa")
print(s.get_name())   # Output: Richa
s.set_name("Anita")
```

```python
Same Example Using @property

class Student:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):          # getter
        return self._name

    @name.setter
    def name(self, value):   # setter
        if len(value) > 0:
            self._name = value
        else:
            print("Name cannot be empty")

# Using property like an attribute
s = Student("Richa")
print(s.name)      # calls getter → Output: Richa

s.name = "Anita"   # calls setter
print(s.name)      # Output: Anita
```

2. Write a program to create an abstract class shape with the method area() derive class's circle and rectangle from it, and implement the area() method in both.

```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass  # Abstract method, must be implemented by derived classes

# Derived class for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Derived class for Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Example usage
circle = Circle(5)
print(f"Area of Circle: {circle.area()}")  # Output: Area of Circle: 78.53981633974483

rectangle = Rectangle(4, 6)
print(f"Area of Rectangle: {rectangle.area()}")  # Output: Area of Rectangle: 24
```

3. Implement a multilevel inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

```python
# Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def show_type(self):
        print(f"Vehicle Type: {self.type}")

# Derived class
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)  # Call Vehicle constructor
        self.brand = brand

    def show_brand(self):
        print(f"Car Brand: {self.brand}")

# Further derived class
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)  # Call Car constructor
        self.battery = battery_capacity

    def show_battery(self):
        print(f"Battery Capacity: {self.battery} kWh")

# Example usage
tesla = ElectricCar("Car", "Tesla", 100)
tesla.show_type()
tesla.show_brand()
tesla.show_battery()


```

5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attribute, balance, and methods to deposit, withdraw, and check balance.

```python
# Class demonstrating encapsulation
class BankAccount:
    def __init__(self, initial_balance):
        self.__balance = initial_balance  # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance or invalid amount.")

    # Method to check balance
    def get_balance(self):
        return self.__balance

# Example usage
account = BankAccount(1000)

account.deposit(500)          # Deposited: 500
account.withdraw(200)         # Withdrawn: 200
print(f"Current Balance: {account.get_balance()}")  # Current Balance: 1300

```

Explanation:

  - Balance is a private attribute, not accessible directly from outside the  class.
  - Methods deposit(), withdraw(), and get_balance() provide controlled access to the private attribute.
  - This ensures the internal state of the object is protected and can only be modified safely.

6. Demonstrate runtime polymorphism using a method Play() in a Base Class Instrument. Derive Classes Guitar and Piano that implement their own version of Play().

```python
# Base class
class Instrument:
    def play(self):
        print("Playing an instrument.")

# Derived class Guitar
class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar.")

# Derived class Piano
class Piano(Instrument):
    def play(self):
        print("Playing the piano.")

# Example usage
def perform(instrument):
    instrument.play()  # Calls the overridden method at runtime

# Creating objects
guitar = Guitar()
piano = Piano()

# Runtime polymorphism in action
perform(guitar)
perform(piano)
```

Explanation:
  - Instrument is the base class with a method play().
  - Guitar and Piano override the play() method.
  - The perform() function takes an Instrument object but calls the appropriate play() method at runtime depending on the object type.
  - This demonstrates runtime (dynamic) polymorphism because the method executed is determined when the program runs, not at compile time.

9. Write a class fraction with attributes Numerator and Denominator. Override the str method to display the fraction as "Numerator / Denumerator".

```python
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    # Overriding the __str__ method
    def __str__(self):
        return f"{self.numerator} / {self.denominator}"

# Example usage
f1 = Fraction(3, 4)
f2 = Fraction(5, 8)

print(f1)
print(f2)
```

Explanation:
  - The class Fraction has two attributes: numerator and denominator.
  - The special method __str__() is overridden to define how the object should be displayed when printed.
  - When you use print(f1), Python automatically calls f1.__str__(), showing the fraction in the format "Numerator / Denominator".

10. Demonstrate operator overloading by creating a class vector and overriding the add method to add two vectors.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Overriding __str__ for readable output
    def __str__(self):
        return f"Vector({self.x}, {self.y})"

# Example usage
v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2  # Calls v1.__add__(v2)

print(v1)
print(v2)
print(v3)
```

Explanation:

  - The method __add__() is a special method used to overload the + operator.
  - When v1 + v2 is executed, Python internally calls v1.__add__(v2).
  - The result is a new Vector object whose x and y components are the sums of the corresponding components of v1 and v2.

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # Expecting a list of numeric grades

    def Average_Grade(self):
        if len(self.grades) == 0:
            return 0  # Prevent division by zero
        return sum(self.grades) / len(self.grades)

# Example usage
student1 = Student("Richa", [85, 90, 78, 92, 88])
average = student1.Average_Grade()
print(f"Average grade of {student1.name}: {average:.2f}")
```

14. Create a class Employee with a method Calculate_Salary() that computes the salary based on hours, work, and hourly rate. Create or derive class manager that adds a bonus to the salary.

```python
# Base class
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate salary
    def Calculate_Salary(self):
        salary = self.hours_worked * self.hourly_rate
        return salary


# Derived class
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    # Overriding the method to include bonus
    def Calculate_Salary(self):
        base_salary = super().Calculate_Salary()
        total_salary = base_salary + self.bonus
        return total_salary


# Example usage
emp1 = Employee("Richa", 160, 200)  # 160 hours at ₹200/hour
mgr1 = Manager("Ankit", 160, 200, 10000)  # Manager with ₹10,000 bonus

print(f"{emp1.name}'s Salary: ₹{emp1.Calculate_Salary()}")
print(f"{mgr1.name}'s Salary (with bonus): ₹{mgr1.Calculate_Salary()}")
```

16. Create a class Animal with an abstract method Sound(). Create two derived classes Cow and Sheep that implement the Sound() method.

```python
from abc import ABC, abstractmethod

# Abstract base class
class Animal(ABC):
    @abstractmethod
    def Sound(self):
        pass  # Abstract method (must be implemented in derived classes)

# Derived class Cow
class Cow(Animal):
    def Sound(self):
        print("Cow says: Moo!")

# Derived class Sheep
class Sheep(Animal):
    def Sound(self):
        print("Sheep says: Baa!")

# Example usage
cow = Cow()
sheep = Sheep()

cow.Sound()
sheep.Sound()
```

18.Create a class House with attributes Address and Price. Create a derived class Mansion that adds an attribute number_of_rooms.

```python
# Base class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def display_info(self):
        print(f"Address: {self.address}, Price: ₹{self.price}")

# Derived class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)  # Call base class constructor
        self.number_of_rooms = number_of_rooms

    def display_info(self):
        # Extend the base class display method
        print(f"Address: {self.address}, Price: ₹{self.price}, Rooms: {self.number_of_rooms}")

# Example usage
house1 = House("123 Main St", 5000000)
mansion1 = Mansion("456 Elite Ave", 25000000, 10)

house1.display_info()
mansion1.display_info()
```
